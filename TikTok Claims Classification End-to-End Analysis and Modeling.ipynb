## **Dataset Structure**

This dataset contains information about TikTok videos where a claim or opinion has been made. It includes **19,383 rows**, each representing a unique published video. The dataset captures **claim status**, differentiating between "opinions" (personal beliefs) and "claims" (unsourced or unverified information). It also includes **video metadata**, such as duration, transcription text, and engagement metrics (view count, like count, share count, download count, and comment count). Additionally, it provides details about the **author**, including **verification status** (whether the user is verified) and **ban status** (active, under review, or banned). Each video is uniquely identified by a **video_id**, while the **#** column represents a TikTok-assigned unique number for videos containing a claim or opinion.

| Column Name                | Type  | Description  |
|----------------------------|-------|-------------|
| #                          | int   | TikTok assigned number for video with claim/opinion. |
| claim_status              | obj   | Whether the published video has been identified as an ‚Äúopinion‚Äù or a ‚Äúclaim.‚Äù In this dataset, an ‚Äúopinion‚Äù refers to an individual‚Äôs or group‚Äôs personal belief or thought. A ‚Äúclaim‚Äù refers to information that is either unsourced or from an unverified source. |
| video_id                  | int   | Random identifying number assigned to video upon publication on TikTok. |
| video_duration_sec        | int   | How long the published video is measured in seconds. |
| video_transcription_text  | obj   | Transcribed text of the words spoken in the published video. |
| verified_status           | obj   | Indicates the status of the TikTok user who published the video in terms of their verification, either ‚Äúverified‚Äù or ‚Äúnot verified.‚Äù |
| author_ban_status         | obj   | Indicates the status of the TikTok user who published the video in terms of their permissions: ‚Äúactive,‚Äù ‚Äúunder review‚Äù or ‚Äúbanned.‚Äù |
| video_view_count         | float | The total number of times the published video has been viewed. |
| video_like_count         | float | The total number of times the published video has been liked by other users. |
| video_share_count        | float | The total number of times the published video has been shared by other users. |
| video_download_count     | float | The total number of times the published video has been downloaded by other users. |
| video_comment_count      | float | The total number of comments on the published video. |

#### **Methods for Identifying Outliers**

Statistical Analysis: I will use NumPy functions to compute the mean and median of the data to understand its central tendency and range. Comparing these values will help me assess skewness:

 - If the mean is greater than the median, the data is likely right-skewed, with high-value outliers pulling the mean upward.
 - If the mean is less than the median, the data is likely left-skewed, with low-value outliers pulling the mean downward.
 - If the mean and median are close, the data is symmetric with no significant outliers.
  
Boxplots: To further explore the data distribution and detect outliers, I will create boxplots. Outliers typically fall outside the whiskers, which represent 1.5 times the interquartile range (IQR) from the first and third quartiles.

IQR Rule: Using the IQR rule, I will identify outliers as values below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR, where IQR = Q3 - Q1.

#### **Deciding Whether to Keep or Remove Outliers**  

My decision to **retain, remove, or reassign outliers** depends on the dataset, the nature of the outliers, and my analysis objectives. Here are the general approaches I will consider:  

- **Keeping Outliers:**  
  - If I am conducting **exploratory data analysis (EDA)** or working with a model that is **robust to outliers**, I may retain them.  
  - Outliers can provide **valuable insights** or represent **valid but rare occurrences** (e.g., viral content on TikTok).  

- **Deleting Outliers:**  
  - If outliers result from **data entry errors** or **typos**, I will remove them.  
  - In **machine learning or predictive modeling**, if outliers **bias results**, I will consider removal, but only with caution.  

- **Reassigning Outliers:**  
  - If the dataset is **small** or extreme values would **disrupt analysis**, I may replace them with a **mean, median, or capped value**.  

**Guidelines for Decision-Making**  
I will assess outliers based on their type:  
- **Collective Outliers** (e.g., patterns deviating from the group) may indicate **meaningful trends**.  
- **Contextual Outliers** (e.g., high views during specific events) require **further investigation**.  
- **Point Outliers** (e.g., single anomalies) are often **errors or special cases**.  

If outliers **skew the analysis** or **reduce model accuracy**, I will consider **removal or reassignment**. Otherwise, I will investigate their significance further to determine whether they offer valuable insights.

### **Task 4. Determine outliers**

When building predictive models, outliers can be problematic. For example, if I were predicting a video's view count, extremely high engagement levels might introduce bias. Additionally, some outliers could indicate issues with data collection or recording.  

The primary objective of the TikTok project is to build a model that predicts whether a video is a **claim or an opinion**. My analysis suggests that a video's **engagement level is strongly correlated with its claim status**. Since all values align with expectations for social media engagement‚Äîwhere a small proportion of videos go viral‚Äîthere‚Äôs no reason to suspect errors in data capture.  

However, it's still important to quantify how many data points could be considered outliers. The definition of an outlier varies by project, and **domain expertise is essential** in setting a threshold. A common method for detecting outliers in a **normal distribution** is using the **interquartile range (IQR)** and defining outliers as values **1.5 * IQR above the third quartile (Q3)**.  

Since the **count variables in this dataset are not normally distributed and are heavily right-skewed**, I will modify the threshold by calculating the **median** for each variable and adding **1.5 * IQR**. This approach results in a lower threshold compared to using Q3, making it more appropriate for detecting outliers in this dataset.

**Key Difference Between Median and Q3**

Median (50th percentile): The value that splits the data into two equal halves. It is robust to extreme values (outliers) and provides a better central tendency for skewed distributions.

Q3 (75th percentile): Represents the upper boundary of the middle 50% of the data. In skewed distributions, Q3 may already be influenced by the skew, as it's closer to the long tail.

#### **Handling Outliers in Skewed Data**

Right-skewed data (as in viral metrics):

Extreme outliers (e.g., videos with millions of views) pull the distribution to the right.

Using Q3 + 1.5 √ó IQR increases the threshold significantly because Q3 is already closer to these extreme values in skewed data. This makes the threshold too high, potentially missing many outliers.

Median is more robust:

Median is not affected by extreme outliers, so adding 1.5 √ó IQR to the median creates a lower threshold.

This lower threshold is better for detecting outliers in heavily skewed data because it doesn't let the outliers influence the benchmark for defining what counts as an outlier.

```python
count_cols = [
    'video_view_count',
    'video_like_count',
    'video_share_count',
    'video_download_count',
    'video_comment_count'
]

for column in count_cols:
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    median = data[column].median()
    outlier_threshold = median + 1.5 * iqr

    outlier_count = (data[column] > outlier_threshold).sum()
    print(f'Number of outliers in {column}: {outlier_count}')
```

### **Task 3: Hypothesis testing**

#### **Defining Hypotheses for TikTok Video Engagement Analysis**

The null hypothesis (H‚ÇÄ) and alternative hypothesis (H<sub>a</sub>) are central to hypothesis testing. The null hypothesis suggests that there is no effect or difference in the population, while the alternative hypothesis proposes that there is an effect or difference. In this project, the hypotheses are as follows:

- Null hypothesis (H‚ÇÄ): There is no difference between the mean video_view_count of verified and unverified TikTok accounts.
- Alternative hypothesis (H<sub>a</sub>): There is a difference between the mean video_view_count of verified and unverified TikTok accounts.

The key distinction between these hypotheses is that the null hypothesis represents a statement of no effect or difference, often associated with equality, while the alternative hypothesis challenges the null and reflects the potential difference or effect being tested.

My goal in this step is to conduct a **two-sample t-test** to analyze the relationship between `verified_status` and `video_view_count`. The process follows these steps:  

1. **State the hypotheses**: Define the null hypothesis (H‚ÇÄ) and the alternative hypothesis (H<sub>a</sub>).  
2. **Choose a significance level**: Determine an appropriate alpha level (e.g., 0.05).  
3. **Calculate the p-value**: Perform the two-sample t-test to obtain the p-value.  
4. **Make a decision**: Reject or fail to reject the null hypothesis based on the p-value.  

By following these steps, I will assess whether there is a significant difference in `video_view_count` between verified and non-verified users.

The null and alternative hypotheses for my analysis are as follows:  

- **Null hypothesis (H‚ÇÄ)**: There is no difference in the number of views between TikTok videos posted by verified and unverified accounts. Any observed difference in the sample data is due to chance or sampling variability.  
- **Alternative hypothesis (H<sub>a</sub>)**: There is a difference in the number of views between TikTok videos posted by verified and unverified accounts. Any observed difference in the sample data is due to an actual difference in the corresponding population means.  

In simpler terms, the null hypothesis assumes that verified and unverified accounts receive the same average number of views, while the alternative hypothesis suggests that a significant difference exists between them.

I proceed with a two-sample t-test using a 5% significance level, in alignment with the organization's standard threshold for statistical testing.

```python
not_verified = data[data["verified_status"] == "not verified"]["video_view_count"]
verified = data[data["verified_status"] == "verified"]["video_view_count"]

stats.ttest_ind(a=not_verified, b=verified, equal_var=False)
```

Based on the p-value, which is extremely small and much smaller than the significance level of 5%, I reject the null hypothesis. This indicates that there is a statistically significant difference in the mean video view counts between verified and unverified TikTok accounts. Specifically, unverified accounts have significantly higher mean view counts than verified accounts.

#### **Evaluating Success in the TikTok Claims Classification Project**

The objective is to reduce the backlog of user-reported videos by developing a machine learning model‚Äîprimarily a Random Forest classifier, with other models considered for comparison‚Äîto distinguish between claims and opinions in reported content. Automating this classification enables TikTok to better prioritize videos for human moderation, ensuring that potentially harmful or misleading claims receive appropriate scrutiny, while opinion-based content is deprioritized. This approach supports faster, more efficient content review and improves platform safety.

One of the most suitable evaluation metric for this task is the **F1 score**, as it balances both precision and recall, which are crucial in this context:  

- **False positives (misclassifying opinions as claims)**: This would unfairly reduce the credibility of users posting opinions, potentially leading to unnecessary content moderation and user dissatisfaction.  
- **False negatives (misclassifying claims as opinions)**: This is particularly problematic because claims may contain misinformation, harmful content, or violations of TikTok's terms of service. If such videos bypass moderation, they could spread widely, damage user trust, and harm the platform‚Äôs credibility.  

Since both false positives and false negatives have serious consequences, optimizing the F1 score ensures a balanced trade-off between precision and recall, leading to a more reliable moderation system.  

#### **Ethical Implications and Consequences of Errors in TikTok‚Äôs Claims Classification Model**

The ethical implications of this model revolve around the potential harm caused by misclassification. In this context, recall is a more important metric than precision because failing to identify a claim (i.e., classifying it as an opinion) could allow harmful or misleading content to bypass moderation. This would constitute a false negative (Type II error), which could result in misinformation spreading unchecked.  

On the other hand, if the model mistakenly classifies an opinion as a claim (false positive or Type I error), the consequences are less severe. Such videos would still undergo additional moderation, where human reviewers or subsequent processes would correctly categorize them. The primary concern is ensuring that potentially harmful claim videos are not overlooked, as this could enable the spread of misinformation, hate speech, or other content that violates TikTok‚Äôs terms of service.  

Thus, prioritizing high recall ensures that the system captures as many claims as possible for further review, minimizing the risk of harmful content slipping through undetected.

For instance, using 2-grams or 3-grams, the text is broken down into word combinations like "a cup" or "for a child." This process builds a matrix where each row represents a video and each column represents an n-gram, with values indicating how often each n-gram appears. To reduce dimensionality and focus on the most relevant information, I retain only the **15 most frequent n-grams** as features for modeling.

I set up a `CountVectorizer` object, which converts a collection of text to a matrix of token counts.

```python
count_vec = CountVectorizer(ngram_range=(2, 3),
                            max_features=15,
                            stop_words='english')
```

I fit the `CountVectorizer` to the **training data only**, generating the n-grams based on that subset. This ensures the model is not exposed to any information from the **validation or test sets**, preserving the integrity of the evaluation process and preventing data leakage.

```python
count_vec.fit(X_train['video_transcription_text'])
```

I fit a Random Forest model to the training set and use cross-validation to tune its hyperparameters. During this process, I evaluate performance based on **recall** and select the model that achieves the highest recall score across the folds.

My model is a **random forest classifier**, an ensemble learning technique that combines multiple decision trees to classify observations. Each tree in the forest is a weak learner that makes predictions based on different subsets of features and training data. Within these trees, splits are made based on **decreasing Gini impurity**, ensuring that each node becomes more homogeneous. The final prediction is determined by a **majority vote** across all trees.  

The model's most predictive features were **user engagement metrics**, such as **views, likes, shares, and downloads**, which help distinguish between claims and opinions.
